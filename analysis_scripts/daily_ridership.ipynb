```python
import pandas as pd
df = pd.read_csv('./MTA_Daily_Ridership_Data__Beginning_2020_20241009.csv')
```

```python
# daily ridership -- more than two stdev is auspicious
# less than two is inauspicious
import numpy as np
df['ridership_stdev'] = np.std(df['Subways: Total Estimated Ridership'])
df['mean_ridership'] = np.mean(df['Subways: Total Estimated Ridership'])
```

```python
df['auspicious_ridership_flag'] = np.where(df['Subways: Total Estimated Ridership'] > 2*df['ridership_stdev'] + df['mean_ridership'], 1, 0)
```

```python
df['inauspicious_ridership_flag'] = np.where(df['Subways: Total Estimated Ridership'] <  df['mean_ridership'] - 2*df['ridership_stdev'], 1, 0)
```
